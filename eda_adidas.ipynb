```python
# to compare multiple column 
category=['Retailer','Region','State','City','Product','Sales Method']
for col in category:
    best=df[[col,'Total Sales']].groupby(col).sum()
    best.plot(kind='bar',figsize=(10,7),title='Total sales Performance by '+col,legend=False,colormap='rainbow')
```

```python
category=['Retailer','Region','State','City','Product','Sales Method']
for col in category:
    best=df[[col,'Operating Profit']].groupby(col).sum()
    best.plot(kind='bar',figsize=(10,7),title='Total Profit Performance by '+col,legend=False,colormap='rainbow')
```
